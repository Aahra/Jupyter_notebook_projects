## Making figures with Numpy arrays

Line plot
scatter plot
bar plot
hist
subplots


```python
import numpy as np
```

```python
# create some data
x= np.linspace(0,10,100)
x[:10]
```

```python
#plot the data and create a line plot
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
#scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig,ax=plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
#making a plot from a dictionary
nut_b_prices ={"almond_butter":10,
              "peanut butter":9,
              "cashew butter":12}
fix,ax=plt.subplots()
ax.bar(nut_b_prices.keys(),nut_b_prices.values())
ax.set(title="Nut butter store",
      ylabel="Price" );
```

```python
fig,ax=plt.subplots()
ax.barh(list(nut_b_prices.keys()),list(nut_b_prices.values()));
```

```python
#histogram,make some data then plot
```

```python
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

```python
#two options for subplots
#option1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                      ncols=2,
                                      figsize=(10,5))
ax1.plot(x,x**2)
ax2.scatter(np.random.rand(10),np.random.rand(10))
ax3.bar(nut_b_prices.keys(),nut_b_prices.values());
```

```python
#option2
fig,ax=plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.rand(10),np.random.rand(10));
```
